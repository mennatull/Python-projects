```python
a=[
["#","#","#"],
["#","#","#"],
["#","#","#"]]

def insert_x(a):
    x=input("Please enter the number of x's you want to enter:")
    for i in range (int(x)):
        row=input("row no:")
        cols=input("col no:")
        a[int(row)][int(cols)]="x"
        for i in range(len(a)):
                print(a[i])
def insert_0(a):
    o=input("Please enter the number of o's you want to enter:")
    for i in range (int(o)):
        row=input("row no:")
        cols=input("col no:")
        a[int(row)][int(cols)]="o"
        for i in range(len(a)):
                print(a[i])
insert_x(a)
insert_0(a)
counto = 0
countx = 0
heuristic_o = 0
heuristic_x = 0
# calculating the heuristic by row

for i in range(len(a)):
    counto = 0
    countx = 0
    for j in range(len(a[i])):
        if a[i][j] == "o":
            counto += 1
        elif a[i][j] == "x":
            countx += 1
    if counto > 0 and not countx:
        heuristic_o += 1
    elif countx > 0 and not counto:
        heuristic_x += 1
    elif not countx and not counto:
        heuristic_o += 1
        heuristic_x += 1

print("row o : " + str(heuristic_o))
print("row x : " + str(heuristic_x))
# Calculating the heuristic by column
for i in range(len(a)):
    count_hash_2 = 0
    counto = 0
    countx = 0
    for j in range(len(a[j])):
        if a[j][i] == "o":
            counto += 1
        elif a[j][i] == "x":
            countx += 1
    if counto > 0 and not countx:
        heuristic_o += 1
    elif countx > 0 and not counto:
        heuristic_x += 1
    elif not countx and not counto:
        heuristic_o += 1
        heuristic_x += 1
print("col o : " + str(heuristic_o))
print("col x : " + str(heuristic_x))

# calcculating the heuristic by diag
countero = 0
counterx = 0
for i in range(len(a)):
    for j in range(len(a)):
        if i == j:
            if a[i][j] == "o":
                countero += 1
            elif a[i][j] == "x":
                counterx += 1
if countero and not counterx:
    heuristic_o += 1
elif counterx and not countero:
    heuristic_x += 1
elif not countero and not counterx:
    heuristic_o += 1
    heuristic_x += 1

print("first o : " + str(heuristic_o))
print("first x : " + str(heuristic_x))
# heuristic of the sec diag

n = len(a)
cnt1 = 0
cnt2 = 0
for i in range(n):
    for j in range(n):
        if (i + j) == (n - 1):
            if a[i][j] == "o":
                cnt1 += 1
            elif a[i][j] == "x":
                cnt2 += 1
if cnt1 and not cnt2:
     heuristic_o+= 1
elif cnt2 and not cnt1:
     heuristic_x+= 1
elif not cnt1 and not cnt2:
    heuristic_o+= 1
    heuristic_x+= 1
print("sec o : " + str(heuristic_o))
print("sec x : " + str(heuristic_x))
final_heuristic=heuristic_x-heuristic_o
print(final_heuristic)


```
